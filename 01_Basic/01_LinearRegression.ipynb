```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```
x = np.linspace(0, 2, 20)
w = 2
beta = -1
y = x * w + beta

plt.plot(x, y)
plt.title("w = " + str(w) + ", beta = " + str(beta))
plt.legend(["x*w+beta"])
plt.xlabel('x')
plt.ylabel('y')
```

```
x = np.linspace(0, 2, 20)
beta = np.random.randint(-2, 2, 1)
w = np.random.randint(2, 10, 1) * 0.5
sigma = np.random.randn(x.shape[0],)

y = x * w + beta + sigma

plt.plot(x, y, '.-', markersize=10, linewidth=1)
plt.title("w = " + str(w) + ", beta = " + str(beta))
plt.legend(["x*w+beta+sigma"])
```

```
w_bar = 6
beta_bar = -3
y_bar = x * w_bar + beta_bar

plt.plot(x, y, '.-', markersize=10, linewidth=1)
plt.plot(x, y_bar, 'r')

plt.title("w = " + str(w) + ", beta = " + str(beta) + "\n" +
          "w_bar = " + str(w_bar) + ", beta_bar = " + str(beta_bar))
plt.legend(["x*w+beta+sigma", "x*w_bar+beta_bar"])
```

```
w_bar = 6
beta_bar = -3
y_bar = x * w_bar + beta_bar

plt.plot(x, y, '.-', markersize=10, linewidth=1)
plt.plot(x, y_bar, 'r', linewidth=2)

error = 0
for i in range(y.shape[0]):
  error += np.power(y[i]-y_bar[i], 2)

for i in range(y.shape[0]):
  plt.plot([x[i], x[i]], [y[i], y_bar[i]], 'k-', linewidth=0.5)

plt.title("w = " + str(w) + ", beta = " + str(beta) + "\n" +
          "w_bar = " + str(w_bar) + ", beta_bar = " + str(beta_bar) + "\n"
          "Error = " + str(round(error,3)))
plt.legend(["x*w+beta+sigma", "x*w_bar+beta_bar", "Error"])
```

```
N = x.shape[0]

w_bar = (N * np.sum(x*y) - np.sum(x)*np.sum(y)) / (N * np.sum(np.power(x, 2)) - np.power(np.sum(x),2))
beta_bar = (np.sum(y)-w_bar * np.sum(x)) / N
y_bar = x * w_bar + beta_bar

plt.plot(x, y, '.-', markersize=10, linewidth=1)
plt.plot(x, y_bar, 'r', linewidth=2)

error = 0
for i in range(y.shape[0]):
  error += np.sqrt(np.power(y[i]-y_bar[i], 2))

for i in range(y.shape[0]):
  plt.plot([x[i], x[i]], [y[i], y_bar[i]], 'k-', linewidth=0.5)

plt.title("w = " + str(w) + ", beta = " + str(beta) + "\n" +
          "w_bar = " + str(round(w_bar,3)) + ", beta_bar = " + str(beta_bar) + "\n"
          "Error = " + str(round(error,3)))
plt.legend(["x*w+beta+sigma", "x*w_bar+beta_bar", "Error"])
```

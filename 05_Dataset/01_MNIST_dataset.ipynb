```
%tensorflow_version 2.x

import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt
```

```
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```
plt.figure(figsize=(10, 20))

for i in range(10):
  idx = np.where(y_train == i)
  np.random.shuffle(idx[0])

  for j in range(5):  
    r_idx = idx[0][j]

    plt.subplot(10,5, (i*5)+j+1)
    plt.imshow(x_train[r_idx], cmap='gray')

plt.tight_layout()
plt.show()
```

```
xr_train = x_train.reshape((x_train.shape[0], x_train.shape[1]*x_train.shape[2]))
xr_test = x_test.reshape((x_test.shape[0], x_test.shape[1]*x_test.shape[2]))
```

```
N_ROW = 200

n_img = np.zeros((N_ROW*10, xr_train.shape[1]))

for i in range(10):
  idx = np.argwhere(y_train == i).flatten()
  
  np.random.shuffle(idx)
  
  for j in range(N_ROW):
    n_img[(i*N_ROW)+j,:] = xr_train[idx[j]]
  
plt.figure(figsize=(10, 20))
plt.imshow(n_img, cmap='gray')
plt.show()
```

```
plt.figure(figsize=(10, 5))

for i in range(10):
  idx = np.argwhere(y_train == i).flatten()
  mean_img = np.mean(x_train[idx], axis=0)

  plt.subplot(2,5, (i+1))
  plt.imshow(mean_img, cmap='gray')

plt.tight_layout()
plt.show()
```

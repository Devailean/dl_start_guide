```
%tensorflow_version 2.x

import tensorflow as tf
from tensorflow.keras import datasets

import matplotlib.pyplot as plt
import numpy as np
```

```
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

train_images, test_images = train_images / 255.0, test_images / 255.0
```

```
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 20))

for i in range(10):
  idx = np.argwhere(train_labels == i)[:,0]
  np.random.shuffle(idx)

  for j in range(5):  
    r_idx = idx[j]

    plt.subplot(10,5, (i*5)+j+1)
    plt.imshow(train_images[r_idx], cmap=plt.cm.binary)
    plt.xlabel(class_names[i])

plt.tight_layout()
plt.show()
```

```
xr_train = train_images.reshape((train_images.shape[0], train_images.shape[1]*train_images.shape[2], 3))
xr_test = test_images.reshape((test_images.shape[0], test_images.shape[1]*test_images.shape[2], 3))
```

```
N_ROW = 200

n_img = np.zeros((N_ROW*10, xr_train.shape[1], 3))

for i in range(10):
  idx = np.argwhere(train_labels == i)[:,0]
  
  np.random.shuffle(idx)
  
  for j in range(N_ROW):
    n_img[(i*N_ROW)+j,:,:] = xr_train[idx[j]]
  
plt.figure(figsize=(20, 30))
plt.imshow(n_img, cmap=plt.cm.binary)
plt.show()
```

```
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 5))

for i in range(10):
  idx = np.argwhere(train_labels == i)[:,0]
  mean_img = np.mean(train_images[idx], axis=0)

  plt.subplot(2,5, (i+1))
  plt.imshow(mean_img, cmap=plt.cm.binary)
  plt.xlabel(class_names[i])

plt.tight_layout()
plt.show()
```
